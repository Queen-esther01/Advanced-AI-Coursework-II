```python
def read_docx(file):
    from docx import Document
    from docx.table import Table
    from docx.text.paragraph import Paragraph

    def iter_block_items(parent):
        """
        Yield paragraphs and tables in document order.
        """
        parent_elm = parent.element.body

        for child in parent_elm.iterchildren():
            if child.tag.endswith('}p'):
                yield Paragraph(child, parent)

            elif child.tag.endswith('}tbl'):
                yield Table(child, parent)


    doc = Document(file)

    structured_content = []

    for block in iter_block_items(doc):

        # Paragraphs
        if isinstance(block, Paragraph):
            structured_content.append({
                "type": "paragraph",
                "style": block.style.name,
                "text": block.text
            })

        # Tables
        elif isinstance(block, Table):
            table_data = []

            for row in block.rows:
                table_data.append([
                    cell.text for cell in row.cells
                ])

            structured_content.append({
                "type": "table",
                "rows": table_data
            })


    from pprint import pprint
    pprint(structured_content)
```

```python
def read_pptx(file):
    from pptx import Presentation
    from pptx.enum.shapes import MSO_SHAPE_TYPE

    prs = Presentation(file)

    structured_presentation = []

    for slide_index, slide in enumerate(prs.slides):

        slide_data = {
            "slide_number": slide_index + 1,
            "elements": []
        }

        for shape in slide.shapes:

            element = {
                "shape_type": str(shape.shape_type)
            }

            # Text
            if hasattr(shape, "text"):
                element["type"] = "text"
                element["text"] = shape.text

            # Tables
            elif shape.shape_type == MSO_SHAPE_TYPE.TABLE:
                table_data = []

                table = shape.table

                for row in table.rows:
                    table_data.append([
                        cell.text for cell in row.cells
                    ])

                element["type"] = "table"
                element["rows"] = table_data

            # Pictures
            elif shape.shape_type == MSO_SHAPE_TYPE.PICTURE:
                element["type"] = "image"
                element["image_name"] = shape.name

            # Charts
            elif shape.has_chart:
                chart = shape.chart

                element["type"] = "chart"
                element["chart_type"] = str(chart.chart_type)

            slide_data["elements"].append(element)

        structured_presentation.append(slide_data)


    from pprint import pprint
    pprint(structured_presentation)
```
